## Calcular a Probabilidade de sobrevivência para Homens e Mulheres

```python

import pandas as pd
import seaborn as sns

# Carregando o dataset do Titanic
df = pd.read_csv('titanic.csv')

# Calculando a probabilidade de sobrevivência para homens
prob_homem = df[(df.Survived == 1) & (df.Sex == 'male')].shape[0] / df[df.Sex == 'male'].shape[0]

# Calculando a probabilidade de sobrevivência para mulheres
prob_mulher = df[(df.Survived == 1) & (df.Sex == 'female')].shape[0] / df[df.Sex == 'female'].shape[0]

# Arredondando as probabilidades e multiplicando por 100 para obter em porcentagem
prob_homem = round(prob_homem, 2) * 100
prob_mulher = round(prob_mulher, 2) * 100

# Criando um gráfico de barras utilizando a biblioteca Seaborn
sns.barplot(x=['Homem', 'Mulher'], y=[prob_homem, prob_mulher])
plt.ylabel('Probabilidade de sobrevivência (%)')
plt.title('Probabilidade de sobrevivência por gênero')
plt.show()



# Calculando a probabilidade de sobrevivência para homens
prob_homem = df[(df.Survived == 1) & (df.Sex == 'male')].shape[0] / df[df.Sex == 'male'].shape[0]

# Calculando a probabilidade de sobrevivência para mulheres
prob_mulher = df[(df.Survived == 1) & (df.Sex == 'female')].shape[0] / df[df.Sex == 'female'].shape[0]

print("Probabilidade de sobreviver sendo homem:", round(prob_homem,2))
print("Probabilidade de sobreviver sendo mulher:", round(prob_mulher,2))
```

## Visualizando tabela de Contigência e Calculando o Qui-quadrado

```python
# Tabela de contingência entre as variáveis 'Survived' e 'Sex'
contingency_table = pd.crosstab(df['Survived'], df['Sex'])

# Visualizar a tabela de contingência
print(contingency_table)

# Plotando gráfico de barras para visualizar a relação entre 'Survived' e 'Sex'
sns.countplot(data=df, x='Survived', hue='Sex')

# Calculando o teste do qui-quadrado
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprimir o valor do qui-quadrado e o valor-p

print("\nChi-quadrado é:", chi2)
print("Valor-p é:", p_value)
```

Após rodar o código em anexo, faça um círculo branco no gráfico do conteúdo em anexo, indicando os pontos $(x1, x2)$ tais que $x1^2$ + $x2^2$ $\sim$ 10. Ou seja, pares $(x1, x2)$ cujo valor da f é próximo de 10.

Considere que "próximo" neste caso significa que o valor resultante da f não pode ser maior, em módulo do que 1. Isto é $-1<f(x) - 10 < 1$, ou seja,  $9 <f(x)<11$

<br>
<div>

</div>

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x1, x2):
    return x1**2 + x2**2

def create_meshgrid(x1_lims, x2_lims):
    x1mesh, x2mesh = np.meshgrid(x1_lims, x2_lims)
    return np.c_[x1mesh.ravel(), x2mesh.ravel()]

# Definindo o tamanho dos vetores para obter uma malha contínua:
x1_lims = np.linspace(-10, 10, 1000)
x2_lims = np.linspace(-10, 10, 1000)

# Criando a malha e usando os vetores definidos acima:
mesh = create_meshgrid(x1_lims, x2_lims)

# Calculando os valores da função f(x1, x2) para todos os pontos da malha
f_values = f(mesh[:, 0], mesh[:, 1])

# Encontrando os índices dos pontos cujo valor da função f(x1, x2) está entre 9 e 11
indices = np.where((f_values > 9) & (f_values < 11))[0]

# Extraindo os pontos (x1, x2) correspondentes aos índices encontrados
x1_selected = mesh[indices, 0]
x2_selected = mesh[indices, 1]

# Plotagem do gráfico
plt.figure(figsize=[15, 10])
plt.scatter(mesh[:, 0], mesh[:, 1], c=f_values, cmap=plt.cm.rainbow)

# Adicionando os círculos brancos nos pontos selecionados
for x1, x2 in zip(x1_selected, x2_selected):
    circle = plt.Circle((x1, x2), 0.2, color='white', fill=False)
    plt.gca().add_patch(circle)

plt.colorbar()
plt.show()
```

### 3 - O tempo de resposta tem alguma correlação com a nota do consumidor?


```python
#importando bibliotecas necessárias
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
```

#### Verificar a correlação entre o tempo de resposta e a nota atribuída pelo consumidor.

```python
#Verificando o tipo de variáveis
print(df.dtypes)
```

####   Verificaricando qual o tipo das duas variáveis para determinar o método estatístico

```python

# Verificando os tipos de dados das colunas de interesse
if 'Tempo Resposta' in df.columns and 'Nota do Consumidor' in df.columns:
    if df['Tempo Resposta'].dtype in ['float64', 'int64'] and df['Nota do Consumidor'].dtype in ['float64', 'int64']:
        # Removendo as linhas com valores ausentes nas colunas de interesse
        df_subset = df[['Tempo Resposta', 'Nota do Consumidor']].dropna()

        # Calculando a correlação entre as duas variáveis
        correlation, p_value = pearsonr(df_subset['Tempo Resposta'], df_subset['Nota do Consumidor'])

        # Imprimindo a correlação e o valor -p
        print("Correlação:", correlation)
        print("Valor-p:", p_value)
    else:
        print("As colunas 'Tempo Resposta' e/ou 'Nota do Consumidor' não são numéricas.")
else:
    print("As colunas 'Tempo Resposta' e/ou 'Nota do Consumidor' não foram encontradas no conjunto de dados.")

```

```python
# Verificando correlação entre o tempo de resposta e a nota atribuída pelo consumidor

df['Data Abertura'] = pd.to_datetime(df['Data Abertura'], dayfirst=True)
df['Ano Abertura'] = df['Data Abertura'].dt.year
df['Mês Abertura'] = df['Data Abertura'].dt.month
df['Dia Abertura'] = df['Data Abertura'].dt.day

#Correlação entre o dia de abertura e a nota do consumidor
df.plot.scatter(x='Dia Abertura', y='Nota do Consumidor')
plt.xlabel('Dia de Abertura')
plt.ylabel('Nota do Consumidor')
plt.title('Correlação entre Dia de Abertura e Nota do Consumidor')
plt.show()

#Correlação entre o mês de abertura e a nota do consumidor
df.plot.scatter(x='Mês Abertura', y='Nota do Consumidor')
plt.xlabel('Mês de Abertura')
plt.ylabel('Nota do Consumidor')
plt.title('Correlação entre Mês de Abertura e Nota do Consumidor')
plt.show()

# Correlação entre ano de abertura e a nota do consumidor
df.plot.scatter(x='Ano Abertura', y='Nota do Consumidor')
plt.xlabel('Ano de Abertura')
plt.ylabel('Nota do Consumidor')
plt.title('Correlação entre Ano de Abertura e Nota do Consumidor')
plt.show()

```

#### Pergunta 3: Existe uma relação entre o segmento de mercado e o grupo de problemas nas reclamações, considerando as colunas 'Segmento de Mercado' e 'Grupo Problema'?

```python
import matplotlib.pyplot as plt

# Agrupar as reclamações por segmento de mercado e grupo de problemas
df_grouped = df.groupby(['Segmento de Mercado', 'Grupo Problema']).size().unstack()

# Plotar o gráfico de barras empilhadas
df_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Segmento de Mercado')
plt.ylabel('Quantidade de Reclamações')
plt.title('Relação entre Segmento de Mercado e Grupo de Problemas')
plt.legend(title='Grupo Problema', bbox_to_anchor=(1, 1))
plt.show()
```
